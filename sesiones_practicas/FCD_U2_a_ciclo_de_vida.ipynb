*An√°lisis exploratorio de las variables de inter√©s (Tipo de propiedad y precio)*

```python
# Calcular cada m√©trica
listings['price'] = listings['price'].replace('[\$,]', '', regex=True).astype(float)

tipos = listings['property_type'].value_counts(normalize=True).mul(100).round(1)

max_precio_por_tipo = listings.groupby('property_type')['price'].max()
min_precio_por_tipo = listings.groupby('property_type')['price'].min()
median_precio_por_tipo = listings.groupby('property_type')['price'].median()
std_precio_por_tipo = listings.groupby('property_type')['price'].std()

# Crear un DataFrame consolidado
tabla_propiedades = pd.DataFrame({
    'Porcentaje (%)': tipos,
    'Cantidad': listings['property_type'].value_counts(),
    'Precio M√≠nimo': min_precio_por_tipo,
    'Precio M√°ximo': max_precio_por_tipo,
    'Precio Mediano': median_precio_por_tipo,
    'Desviaci√≥n Est√°ndar': std_precio_por_tipo
}).sort_values(by='Porcentaje (%)', ascending=False)

display(tabla_propiedades)
```

```python
# --- C√°lculo de estad√≠sticas clave ---
n_listings = len(listings)

tipos = listings['property_type'].value_counts(normalize=True).mul(100).round(1)
tipo1, pt1 = tipos.index[0], tipos.iloc[0]
tipo2, pt2 = tipos.index[1], tipos.iloc[1]

# --- Crear tabla resumen ---
tabla = pd.DataFrame({
    "Categor√≠a": [
        "üì¶ N.¬∫ de registros",
        "üè† Tipo de alojamiento m√°s com√∫n",
    ],
    "Resultado": [
        f"Listings: {n_listings}",
        f"{tipo1} ({pt1}%), seguido de {tipo2} ({pt2}%)"
    ]
})

# Tabla
tabla.style.set_properties(**{'text-align': 'left'})
```

## <span style="color:#2F749F;"><strong>CRISP-DM ‚Äì Fase 3: Preparaci√≥n de los datos </strong></span>

Un proceso de preparaci√≥n adecuado mejora tanto la **precisi√≥n** como la **estabilidad** de los modelos.  
Para reforzar la repetibilidad, se emplean **conductos de datos automatizados** desarrollados en Python, R o mediante herramientas ETL.  


### **Ingenier√≠a de caracter√≠sticas**
Esta tarea combina lo t√©cnico con lo creativo:  
- Integraci√≥n de variables.  
- C√°lculo de retardos o diferencias.  
- Transformaciones espec√≠ficas del dominio de aplicaci√≥n.  

### **Documentaci√≥n y transparencia**
Registrar cada paso de preparaci√≥n es esencial para garantizar **reproducibilidad** y **claridad**, especialmente en contextos sujetos a regulaci√≥n estricta.  

<div style="background-color:#E6F4FF; padding:10px; border-radius:8px;">
<b>AIRBNB</b><br>  
<br> - Limpia y transforma precios.
<br> - Estandariza formatos para evitar errores en modelado o visualizaci√≥n.
<br> - Eliminaci√≥n de outliers con IQR

```python
# Convertir precios a num√©ricos
listings_interest['price'] = pd.to_numeric(listings_interest['price'].replace('[\$,]', '', regex=True), errors='coerce')

# Funci√≥n para eliminar outliers por grupo usando IQR
def remove_outliers_iqr(group):
    q1 = group['price'].quantile(0.25)
    q3 = group['price'].quantile(0.75)
    iqr = q3 - q1
    lim_inf = q1 - 1.5 * iqr
    lim_sup = q3 + 1.5 * iqr
    return group[(group['price'] >= lim_inf) & (group['price'] <= lim_sup)]

# Aplicar la funci√≥n por tipo de propiedad
listings_clean = listings_interest.groupby('property_type', group_keys=False).apply(remove_outliers_iqr)

filas_por_propiedad = listings_clean['property_type'].value_counts()

print(filas_por_propiedad)
```

```python
tipos_clean = listings_clean['property_type'].value_counts(normalize=True).mul(100).round(1)

max_precio_por_tipo_clean = listings_clean.groupby('property_type')['price'].max()
min_precio_por_tipo_clean = listings_clean.groupby('property_type')['price'].min()
median_precio_por_tipo_clean = listings_clean.groupby('property_type')['price'].median()
std_precio_por_tipo_clean = listings_clean.groupby('property_type')['price'].std()

# Crear un DataFrame consolidado
tabla_propiedades_clean = pd.DataFrame({
    'Porcentaje (%)': tipos_clean,
    'Cantidad': listings_clean['property_type'].value_counts(),
    'Precio M√≠nimo': min_precio_por_tipo_clean,
    'Precio M√°ximo': max_precio_por_tipo_clean,
    'Precio Mediano': median_precio_por_tipo_clean,
    'Desviaci√≥n Est√°ndar': std_precio_por_tipo_clean
}).sort_values(by='Porcentaje (%)', ascending=False)

display(tabla_propiedades_clean)
```

```python
df_box_original = pd.DataFrame({
    'Precio': listings_interest['price'],
    'Tipo_propiedad': listings_interest['property_type'],
    'Dataset': 'Con outliers'
})

df_box_clean = pd.DataFrame({
    'Precio': listings_clean['price'],
    'Tipo_propiedad': listings_clean['property_type'],
    'Dataset': 'Sin outliers'
})

# Combinar
df_box_comparado = pd.concat([df_box_original, df_box_clean])

plt.figure(figsize=(14, 6))
sns.boxplot(
    data=df_box_comparado,
    x='Tipo_propiedad',
    y='Precio',
    hue='Dataset',
    palette={'Con outliers': 'salmon', 'Sin outliers': 'lightblue'}
)

plt.title('Comparaci√≥n de precios por tipo de propiedad (Con y Sin Outliers)')
plt.xlabel('Tipo de propiedad')
plt.ylabel('Precio por Noche (USD)')
plt.xticks(rotation=45)
plt.legend(title='Dataset')
plt.tight_layout()
plt.show()
```

```python
tipos = listings_interest['property_type'].unique()

fig, axes = plt.subplots(len(tipos), 2, figsize=(12, 4 * len(tipos)))
fig.suptitle('Histogramas de precios por tipo de propiedad (Con y Sin Outliers)', fontsize=16)

for i, tipo in enumerate(tipos):
    # Histograma con outliers
    sns.histplot(
        data=listings_interest[listings_interest['property_type'] == tipo],
        x='price',
        bins=30,
        ax=axes[i, 0],
        color='salmon',
        kde=True
    )
    axes[i, 0].set_title(f'{tipo} (Con outliers)')
    axes[i, 0].set_xlabel('Precio (USD)')
    axes[i, 0].set_ylabel('Frecuencia')

    # Histograma sin outliers
    sns.histplot(
        data=listings_clean[listings_clean['property_type'] == tipo],
        x='price',
        bins=30,
        ax=axes[i, 1],
        color='lightblue',
        kde=True
    )
    axes[i, 1].set_title(f'{tipo} (Sin outliers)')
    axes[i, 1].set_xlabel('Precio (USD)')
    axes[i, 1].set_ylabel('Frecuencia')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
